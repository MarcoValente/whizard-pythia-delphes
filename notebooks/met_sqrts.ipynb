```python
import re
import os
from pprint import pprint as p
from glob import glob
import lheanalysis as lhea
import pylhe

lhe_filename = '../files/output/test/15TeV/mumu_nunuHH.lhe'

whizard_dirs = sorted(glob('../files/output/whizard/mumu_HH_xs_restr_lhe/*'))
p(whizard_dirs)

values_dict = {
    'MET': {
        'hist_opts' : {'bins':25,'range':(0,1000)},
        'plot_opts': {'xlabel': '$MET (GeV)$'},
        'vals' : [],
    },
    'MEL': {
        'hist_opts' : {'bins':100,'range':(0,10000)},
        'plot_opts': {'xlabel': '$MEL (GeV)$'},
        'vals' : [],
    },
}
```

```python
values_dict_energ = {}
for d in whizard_dirs:
    energy=os.path.basename(d)
    if not energy in values_dict_energ.keys():
        import copy
        values_dict_energ[energy] = copy.deepcopy(values_dict)
    lhe_filename=d+'/mumu_nunuHH.lhe'
    lhea.fillHistograms(values_dict_energ[energy],lhe_filename)
values_dict_energ
```

```python
def plotHistogram(vals,hist_dict,plot_dict):
    import matplotlib.pyplot as plt
    %matplotlib widget

    fig = plt.figure()

    plt.hist(vals, alpha=0.5, **hist_dict)
    # plt.hist(vals)

    if 'logy'   in plot_dict: plt.semilogy(plot_dict['logy'])
    if 'xlabel' in plot_dict: plt.xlabel(plot_dict['xlabel'])
    if 'ylabel' in plot_dict: plt.ylabel(plot_dict['ylabel'])
    if 'title'  in plot_dict: plt.ylabel(plot_dict['title'])
    # plt.legend(loc='best', ncol=2)

    plt.show()
```

```python
hname = 'MEL'
energy = '30.0TeV'
plotHistogram(values_dict_energ[energy][f'{hname}']['vals'],
              values_dict_energ[energy][f'{hname}']['hist_opts'],
              values_dict_energ[energy][f'{hname}']['plot_opts'],
              )
```
