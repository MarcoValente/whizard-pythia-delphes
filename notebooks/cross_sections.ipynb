```python
from glob import glob
from pprint import pprint as p
import os

input_dir='/afs/cern.ch/user/v/valentem/MuonCollider/grid_outputs/whizard/10TeV'

stdout_files=glob(input_dir+'/*.log/tarball*/payload.stdout')
sample_names = set([os.path.basename(os.path.dirname(os.path.dirname(stdout))) for stdout in stdout_files])

print('stdout_files =')
p(stdout_files)
print('sample_names =')
p(sample_names)
```

```python
def extractCrossSections(fname,regex=r'^\s+[\dE+-]+\s+[\dE+-].*'):
    # print(f'Extracting cross section from file {fname}')
    import re
    with open(fname) as f:
        text = f.read().strip()
        m = re.findall(regex, text, re.MULTILINE)
    
    if len(m)==0: return None,None
    xs_string = m[-1].strip()
    xs,xs_e=xs_string.split()[2:4]
    return float(xs),float(xs_e)

xs_dict = {}
for sn in sample_names:
    print(sn)
    for stdout in stdout_files:
        if sn not in stdout: continue
        _xs_tuple = extractCrossSections(stdout)
        if _xs_tuple[0] is None: continue
        if sn not in xs_dict.keys():
            xs_dict[sn] = []
        xs_dict[sn] += [_xs_tuple]
```

```python
import numpy as np

final_xs_dict = {}
for sample_name, xs in xs_dict.items():
    p(sample_name)
    xs_np = np.array(xs)
    xs = xs_np[:,0]
    xs_e = xs_np[:,1]
    xs_mean = np.average(xs)
    xs_e_mean = np.average(xs_e)
    pull = (xs - xs_mean)/xs_e
    print(f'xs = {xs_mean} +- {xs_e_mean} (pull_mean = {np.mean(pull)}, pull_std = {np.std(pull)}')
    final_xs_dict[sample_name.split('.')[3]] = {'xs':xs_mean,'xs_err':xs_e_mean}
```
